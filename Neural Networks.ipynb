Q2: Part 1 : Importing the data

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import roc_auc_score
import numpy as np
train_X=pd.read_csv("trnX-3d.csv",header=None,sep='\t')
dev_X=pd.read_csv("devX-3d.csv",header=None,sep='\t')
train_Y=pd.read_csv("trnY-3d.csv",header=None,sep='\t')
dev_Y=pd.read_csv("devY-3d.csv",header=None,sep='\t')
```

Fitting the train data

```python
classifier1 = MLPClassifier(max_iter=2000)
classifier1.fit(train_X,np.ravel(train_Y))
```

Accuracy

```python
y_pred = classifier1.predict(train_X)
print("Train Accuracy with default params :"+accuracy_score(train_Y, y_pred))
```

Fitting the dev data

```python
classifier2 = MLPClassifier(max_iter=3000)
classifier2.fit(dev_X,np.ravel(dev_Y))
```

Accuracy

```python
y_pred = classifier1.predict(dev_X)
print("Dev Accuracy with default params"+accuracy_score(dev_Y, y_pred))
```

# Question-2 : PART B

```python
classifier1 = MLPClassifier(hidden_layer_sizes=(100,),activation='relu',solver='adam',learning_rate_init=0.001,batch_size='auto',max_iter=5000)
```

```python
classifier1.fit(train_X,np.ravel(train_Y))
```

```python
y_pred = classifier1.predict(train_X)
print("Accuracy after using 5 default params :",accuracy_score(train_Y, y_pred))
```

```python
parameters = {
    'hidden_layer_sizes': [(100,), (20, 40), (50, 50)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'learning_rate_init': [0.001,0.01, 0.1],
    'batch_size': [32,48,64]
}
grid_search_dt = GridSearchCV(estimator = classifier1,
param_grid = parameters,
scoring = 'accuracy', 
cv = 4,
verbose = 2)

#fit the data on 
grid_search_dt.fit(train_X, np.ravel(train_Y))


print(grid_search_dt.best_estimator_)
print(grid_search_dt.best_params_)

# predicting accuracy based on the best params
grid_predictions = grid_search_dt.predict(train_X)

print(classification_report(train_Y, grid_predictions))
score = roc_auc_score(train_Y, grid_predictions)
print(f"ROC AUC: {score:.4f}")
```
